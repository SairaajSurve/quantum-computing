# Gate Error in Quantum Circuit
$<\Psi |$   $=$  $X^d |0>$  
$<Z>$   $=$   $<\Psi |Z|\Psi >$   =   $cos(d\pi + d\epsilon)$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['science', 'notebook', 'grid'])
```

```python
epsilon = 3 # degree
epsilon_radians = (epsilon*np.pi)/180 # degree
```

```python
d = np.arange(0, 101)
```

```python
cos = np.cos(d*(np.pi + epsilon_radians)) # theta in radians
```

```python
plt.plot(d, cos, 'o-' , color='grey', markerfacecolor='red')
plt.ylabel('Measured <Z>')
plt.xlabel('d')
plt.show()
```
