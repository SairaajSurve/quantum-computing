```python
from qiskit import *
```

```python
circuit = QuantumCircuit(3, 3)
```

```python
# circuit.x(0) # change state from |0> to |1>
circuit.h(0)
circuit.barrier()
```

```python
circuit.h(1)
```

```python
circuit.cx(1, 2)
```

```python
circuit.barrier()
```

```python
circuit.cx(0, 1)
```

```python
circuit.h(0)
```

```python
circuit.measure(1, 1)
```

```python
circuit.measure(0, 0)
```

```python
circuit.cx(1, 2)
```

```python
circuit.cz(0, 2)
```

```python
circuit.measure(2,2)
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shots=1024).result()
from qiskit.tools.visualization import plot_histogram
plot_histogram(result.get_counts(circuit))
```

### Thus state of first qbuit is teleported to third qubit

```python
circuit.draw(initial_state=True, output='mpl', cregbundle=False)
```
